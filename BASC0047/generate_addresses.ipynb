```python
import requests
import pandas as pd
import random

def get_first_part(address):
    return ",".join(address.split(",")[:-1])

df = pd.read_csv("fake_addresses.csv")
street_names = df['Address'].apply(get_first_part).tolist()

print(street_names)
```

```python
df = pd.DataFrame() 
entries = 200

i = 0
while i < entries:
    response = requests.get('https://api.postcodes.io/random/postcodes')
    assert response.status_code == 200
    data = response.json()
    assert data['status'] == 200
    result = data['result']

    address_line_1 = street_names[i]
    city = result['admin_district'] # Not perfect, but best thing to use
    postcode = result['postcode']
    latitude = result['latitude']
    longitude = result['longitude']

    if not latitude or not longitude:
        continue
    
    row = {"address_line_1": address_line_1,
           "city": city,
           "postcode": postcode,
           "latitude": latitude,
           "longitude": longitude}

    df = pd.concat([df, pd.DataFrame([row])], ignore_index=True)
    print(row)
    i += 1

print()
print(df.head())
assert len(df) == entries
```

```python
df.to_csv('full_addresses.csv', index=False)
```
